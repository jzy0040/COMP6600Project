### Calculate P(Y=y) for all possible y

```python
def calculate_prior(df, Y):
    classes = sorted(list(df[Y].unique()))
    prior = []
    for i in classes:
        prior.append(len(df[df[Y]==i])/len(df))
    return prior
```

### Approach 1: Calculate P(X=x|Y=y) using Gaussian dist.

```python
def calculate_likelihood_gaussian(df, feat_name, feat_val, Y, label):
    feat = list(df.columns)
    df = df[df[Y] == label]
    mean, std = df[feat_name].mean(), df[feat_name].std()
    p_x_given_y = (1 / (np.sqrt(2 * np.pi) * std)) *  np.exp(-((feat_val-mean)**2 / (2 * std**2 )))
    return p_x_given_y
```

### Calculate P(X=x1|Y=y)P(X=x2|Y=y)...P(X=xn|Y=y) * P(Y=y) for all y and find the maximum

```python
def naive_bayes_gaussian(df, X, Y):
    # get feature names
    features = list(df.columns)[1:-1]

    # calculate prior
    prior = calculate_prior(df, Y)

    Y_pred = []
    # loop over every data sample
    for x in X:
        # calculate likelihood
        labels = sorted(list(df[Y].unique()))
        likelihood = [1]*len(labels)
        for j in range(len(labels)):
            for i in range(len(features)):
                likelihood[j] *= calculate_likelihood_gaussian(df, features[i], x[i], Y, labels[j])

        # calculate posterior probability (numerator only)
        post_prob = [1]*len(labels)
        for j in range(len(labels)):
            post_prob[j] = likelihood[j] * prior[j]

        Y_pred.append(np.argmax(post_prob))

    return np.array(Y_pred)
```

### Test Gaussian model

```python
from sklearn.model_selection import train_test_split

data['diagnosis'].replace('M', 1, inplace = True)
data['diagnosis'].replace('B', 0, inplace = True)
train, test = train_test_split(data, test_size = 0.2, random_state = 41)

X_test = test.iloc[:, 1: -1].values
Y_test = test.iloc[:, 0].values

Y_pred = naive_bayes_gaussian(train, X = X_test, Y = "diagnosis")
```

```python
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score
print(confusion_matrix(Y_test, Y_pred))
print(accuracy_score(Y_test, Y_pred))
print(f1_score(Y_test, Y_pred))
```

```python
size = []
acc = []
for i in range(1, train.shape[0]):
    size.append(i)
    Y_pred = naive_bayes_gaussian(train[: i], X = X_test, Y = "diagnosis")
    acc.append(accuracy_score(Y_test, Y_pred))
```

```python
def plot_learningCurve(size, acc):
    # Plot training & validation accuracy values
    plt.plot(size, acc)
    plt.title('NBC learning curve')
    plt.xlabel('Train examples')
    plt.ylim([0.3, 1])
    plt.ylabel('Accuracy')
    plt.xlim([1, 455])
    plt.show()

plot_learningCurve(size, acc)
```

The learning rate of the Naive bayes classifier is the accuracy with respect to the number of the training samples. We can see that the accuracy is positively correlate to the number training samples.
